2. Transform the 'datetime' column into a datetime type and use it to create 3 new columns in the data frame containing the:
* Name of the Month
* Name of the Day of the Week
* Hour of the Day

```python
# lets convert datetime column to date and time
df['datetime'] = pd.to_datetime(df['datetime'])
df.info()
```

```python
# lets create new columns "month (name)", "week (name)" and "hours of day"
df['month (name)'] = df['datetime'].dt.month_name()
df['week (name)'] = df['datetime'].dt.day_name()
df['hours of day'] = df['datetime'].dt.hour
df.head()
```

```python
# lets create function where our time will be split in "Early Morning", "Morning", "Eve", "Night" and "Late Night"
def f(x):
    if (x > 4) and (x <= 8):
        return 'Early Morning'
    elif (x > 8) and (x <= 12 ):
        return 'Morning'
    elif (x > 12) and (x <= 16):
        return'Noon'
    elif (x > 16) and (x <= 20) :
        return 'Eve'
    elif (x > 20) and (x <= 24):
        return'Night'
    elif (x <= 4):
        return'Late Night'
```

```python
# now lets apply our function from above using "apply()"
df['hours of day'] = df['hours of day'].apply(f) 
```

```python
# lets check if our function worked or not
df['hours of day'].value_counts()
```

```python
# Drop the 'datetime' and 'season' columns. These are now redundant.
df.drop(columns=['datetime', 'season'], inplace=True)
df.head()
```

###3. The temperatures in the 'temp' and 'atemp' columns are in Celsius. Use `.apply()` and a Lambda function to convert

```python
# lets convert "temp" and "atemp" column from Celsius to Fahrenheit
df['temp'] = df['temp'].apply(lambda x: (x*(9/5))+32)
df['atemp'] = df['atemp'].apply(lambda x: (x*(9/5))+32)
```

```python
df.info()
```

```python
# Create a new column, 'temp_variance' which shows how much warmer or colder the current temperature ('temp') is than the average temperate for that day of the year ('atemp'). 
df['temp_variance'] = df['temp'] - df['atemp']
df.head()
```

```python
# Drop the 'atemp' column.
df.drop(columns='atemp', inplace=True)
df.head()
```
